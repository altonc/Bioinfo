```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#target for heart disease protein -- CHEMBL3344
target = new_client.target
target_query = target.search('heart')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target_id = targets.target_chembl_id[6]
selected_target_id
```

```python
activity = new_client.activity
print(selected_target_id)
result = activity.filter(target_chembl_id=selected_target_id).filter(standard_type='IC50')
```

```python
df = pd.DataFrame.from_dict(result)
df.head()
#print(df['standard_value'].sort_values().head())
```

```python
df.to_csv('bioactivity_data.csv', index=False)
```

```python
#clean NA values
df_clean = df[df.standard_value.notna()]
df_clean
```

```python
#preprocess data, label them as either active, inactive or intermediate
bioactivity_class = []
for i in df_clean.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append('inactive')
  elif float(i) <= 1000:
    bioactivity_class.append('active')
  else:
    bioactivity_class.append('intermediate')
```

```python
#filter out molecule_chembl_id's, canonical smiles, and standard values so only one shows up
filter = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df_clean[filter]
df3 = pd.concat([df3, pd.Series(bioactivity_class, name = 'Class')], axis=1)
```
